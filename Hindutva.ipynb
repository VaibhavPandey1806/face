```python
# this code will work in idle python
# error has been debugged so code is error free
# just download libraries numpy and cv2 using comman line interface
# python
# https://github.com/Itseez/o...
"""https://github.com/Itseez/o...

"""
# download this two .xml files in current dir you are working
# here we go

import numpy as np
import cv2

# multiple cascades: https://github.com/Itseez/o...

#https://github.com/Itseez/o...
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
#https://github.com/Itseez/o...
eye_cascade = cv2.CascadeClassifier('haarcascade_eye.xml')

cap = cv2.VideoCapture(0)

while 1:
 _, img = cap.read()
 gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
 faces = face_cascade.detectMultiScale(gray,1.3,5)
 for (x,y,w,h) in faces:
  cv2.rectangle(img,(x,y),(x+w,y+h),(200,0,0),2)
  cv2.rectangle(img,(x,y),(x+w,y+h),(300,2,2),3)
 roi_gray = gray[y:y+h, x:x+w]
 roi_color = img[y:y+h, x:x+w]

 eyes = eye_cascade.detectMultiScale(roi_gray)
 for (ex,ey,ew,eh) in eyes:
  cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,255,0),2)
  cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(2,300,2),3)

  cv2.imshow('img',img)
  k = cv2.waitKey(30) & 0xff
  if k == 27:
   break

cap.release()
cv2.destroyAllWindows()
```
